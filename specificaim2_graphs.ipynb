###Training Curves

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
data = pd.read_csv('/content/drive/MyDrive/sa2_python_trainingcurve.csv')

# Extract data for each column
iterations = data['Iteration']
loss = data['loss']
scmap_loss = data['scmap_loss']
locref_loss = data['locref_loss']
limb_loss = data['limb_loss']

# Create a scatter plot
plt.figure(figsize=(10, 6))  # Adjust size as needed

# Plot each series
plt.scatter(iterations, loss, label='loss')
plt.scatter(iterations, scmap_loss, label='scmap_loss')
plt.scatter(iterations, locref_loss, label='locref_loss')
plt.scatter(iterations, limb_loss, label='limb_loss')

# Add labels and legend
plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.legend()

# Save the plot as PNG
plt.savefig('plot.png', bbox_inches='tight')

# Save the plot as SVG
plt.savefig('plot.svg', format='svg', bbox_inches='tight')

# Show the plot
plt.show()
```

###mean average Euclidean error of bodyparts

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV file
data = pd.read_csv('/content/drive/MyDrive/sa2_python_bodypart_rmse.csv')

# Extract data for plotting
bodyparts = data['bodypart']
train_values = data['train']
test_values = data['test']

# Set the positions for the bars
x = np.arange(len(bodyparts))  # Adjusted to create space between bars

# Define the width of the bars
bar_width = 0.35  # Adjust as needed

# Create bar plots
plt.figure(figsize=(10, 6))  # Adjust size as needed

plt.bar(x - bar_width/2, train_values, color='blue', width=bar_width, label='Train')
plt.bar(x + bar_width/2, test_values, color='orange', width=bar_width, label='Test')

# Add labels and legend
plt.ylabel('Mean Average Euclidean Error (pixels)')
plt.xticks(x, bodyparts, rotation=90)  # Rotate x-axis labels by 90 degrees
plt.legend()

# Save the plot as PNG
plt.savefig('bar_plot.png', bbox_inches='tight')

# Save the plot as SVG
plt.savefig('bar_plot.svg', format='svg', bbox_inches='tight')

# Show the plot
plt.show()
```

###Butts behaviour classifier performance metrics

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/sa2_python_butts.csv')

Parameters1 = data['parameter1'].iloc[[0, 2, 4]].tolist()
Parameters2 = data['parameter2'].iloc[[0, 2, 4]].tolist()
Parameters3 = data['parameter3'].iloc[[0, 2, 4]].tolist()
Parameters1_yerr = data['parameter1'].iloc[[1, 3, 5]].tolist()
Parameters2_yerr = data['parameter2'].iloc[[1, 3, 5]].tolist()
Parameters3_yerr = data['parameter3'].iloc[[1, 3, 5]].tolist()

barWidth = 0.20  # Decreased bar width to add gaps between groups
gap = 0      # Width of the gap between groups

br1 = np.arange(len(Parameters1))
br2 = [x + barWidth + gap for x in br1]
br3 = [x + barWidth + gap for x in br2]

plt.bar(br1, Parameters1, color ='r', width = barWidth,
        edgecolor ='grey', label ='Parameters1', yerr=Parameters1_yerr)
plt.bar(br2, Parameters2, color ='g', width = barWidth,
        edgecolor ='grey', label ='Parameters2', yerr=Parameters2_yerr)
plt.bar(br3, Parameters3, color ='b', width = barWidth,
        edgecolor ='grey', label ='Parameters3', yerr=Parameters3_yerr)

plt.ylabel('Performance')
plt.xticks([r + barWidth + 0.1 for r in range(len(Parameters1))],
        ['Recall', 'Precision', 'F1'])
plt.legend()

# Set the y-axis limit to 1.0
plt.ylim(0, 1.0)

# Save the plot as a vector file
plt.savefig('performance_plot.svg', format='svg')

# Show the plot
plt.show()
```

###Average speed of each body part

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from CSV
data = pd.read_csv('/content/drive/MyDrive/sa2_python_speed_average.csv')

# Extract body parts, averages, and standard errors
body_parts = data['bodypart']
averages = data['average']
std_errors = data['SEM']

# Plot the bar plot
plt.figure(figsize=(10, 6))
plt.bar(body_parts, averages, color='maroon', yerr=std_errors, capsize=5)

# Labeling the axes and title
plt.ylabel("Average speed (pixels/second)")

# Rotate x-axis labels for better visibility
plt.xticks(rotation=90, ha='right')

# Save the plot as a vector file
plt.savefig('average_speed.svg', format='svg', bbox_inches='tight')

# Show the plot
plt.tight_layout()
plt.show()
```

###Speed across frames

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load Data
df = pd.read_csv('/content/drive/MyDrive/BEHAVIOMICS/SpecificAim2/202405_visualisations/sa2_python_speed.csv')

# Step 2: Prepare Data
frames = df['frame']
speed_body_part1 = df['dorsal_mid']
speed_body_part2 = df['dorsal_opercular_mid']
speed_body_part3 = df['ventral_opercular']

# Step 3: Plotting
plt.figure(figsize=(10, 6))

plt.scatter(frames, speed_body_part1, label='dorsal_mid', color='blue', marker='o')
plt.scatter(frames, speed_body_part2, label='dorsal_opercular_mid', color='red', marker='s')
plt.scatter(frames, speed_body_part3, label='ventral_opercular', color='green', marker='^')

plt.xlabel('Frames')
plt.ylabel('Speed (pixels/second)')

plt.legend()
plt.tight_layout()

# Save & Show the plot
plt.savefig('speed_frame.svg', format='svg', bbox_inches='tight')
plt.show()
```
